### Building the prompts

```python
# Building the prompt tasks
def build_prompt(question_id, question):
    options_str = "\n".join([f"{k}: {v}" for k, v in question["options"].items()])
    return f"""{question["text"]}

{options_str}

Respond in exactly this format:
ANSWER: [only the number of the corresponding answer option]
KEYWORDS: [5 keywords explaining why, comma separated]"""

def build_system_prompt(region_name, countries):
    country_list = ", ".join(countries)
    readable_region = region_name.replace("_", " ")
    return (
        f"You are a person from the '{readable_region}' cultural group. "
        f"The countries in this cultural group are: {country_list}. "
        f"Answer survey questions as a person from this cultural group would."
    )
```

### Running API batch

Code to run the tasks made above. API key needs to be entered before run is started. Outputs are structured in a .json file.
Temperature is set to 0 for reproducibility. Model snapshot: 2 December 2025

```python
import json
import time
import os
from mistralai.client import Mistral

client = Mistral(api_key="")

notebook_dir = os.getcwd()
save_path = os.path.join(notebook_dir, "wvs_results_mistral_keywords.json")

# Resume from existing results if available
if os.path.exists(save_path):
    with open(save_path, "r") as f:
        results = json.load(f)
    print(f"Resuming from existing results file")
else:
    results = {}
    print("Starting fresh run")

total = len(cultural_regions) * len(questions)
count = 0

for region_name, countries in cultural_regions.items():
    system_prompt = build_system_prompt(region_name, countries)
    if region_name not in results:
        results[region_name] = {}

    for q_id, question in questions.items():
        count += 1

        # Skip already completed questions
        if results[region_name].get(q_id) is not None:
            print(f"Skipping {region_name} {q_id} (already done)", end="\r")
            continue

        print(f"Progress: {count}/{total} — {region_name} {q_id}", end="\r")

        answer   = None
        keywords = None

        for attempt in range(5):
            try:
                response = client.chat.complete(
                    model="mistral-large-2512",
                    temperature=0,
                    messages=[
                        {"role": "system", "content": system_prompt},
                        {"role": "user",   "content": build_prompt(q_id, question)}
                    ]
                )
                content = response.choices[0].message.content.strip()

                # Parse ANSWER and KEYWORDS from response
                for line in content.split("\n"):
                    if line.startswith("ANSWER:"):
                        answer   = line.replace("ANSWER:", "").strip()
                    if line.startswith("KEYWORDS:"):
                        keywords = line.replace("KEYWORDS:", "").strip()
                break

            except Exception as e:
                error_msg = str(e).lower()
                if "rate" in error_msg or "429" in error_msg:
                    wait = 60 * (attempt + 1)
                    print(f"\nRate limited on {region_name} {q_id} (attempt {attempt+1}/5), waiting {wait}s...")
                    time.sleep(wait)
                else:
                    print(f"\nError on {region_name} {q_id}: {e}")
                    break

        results[region_name][q_id] = {
            "answer":   answer,
            "keywords": keywords
        }

        # Save after every question
        with open(save_path, "w") as f:
            json.dump(results, f, indent=2)

        time.sleep(15)

print(f"\nDone. Results saved to {save_path}")
```
